# Gráficos de curvas paramétricas

```python
cardioid = (x**2 + y**2 - 2 * x)**2 - 4 * (x**2 + y**2)
cardioid
```

```python
u = sym.symbols('u')
sym.plot_parametric((sym.cos(u), sym.sin(u)), (u, -5, 5), size = (3,3))
```

```python
sym.plot_parametric((2*sym.cos(u) - sym.cos(2*u), 2 * sym.sin(u) - sym.sin(2*u)), (u, -5, 5), size = (3,3))
```

```python
t = np.linspace(0,2*np.pi,100)
x = 2 * np.cos(t) - np.cos(2*t)
y = 2 * np.sin(t) - np.sin(2*t)
# Plot line with RGB tuple (red=1, green=0.2, blue=0.5)
plt.figure(figsize=(4, 4))
plt.plot(x, y, label='cardioid curve')
plt.axis()
plt.grid(True)

plt.show()
```

```python
t = np.linspace(0,2*np.pi,100)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.figure(figsize=(6, 4))
# Plot line with RGB tuple (red=1, green=0.2, blue=0.5)
plt.plot(x,y,c=(1,0.2,0.5),lw=20)
# Add features to our figure
plt.title('Heart!')
plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
#
theta = np.linspace(-4*np.pi, 4*np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
#
plt.rcParams['legend.fontsize'] = 10
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='parametric curve')
ax.legend()
plt.show()
```

# Cálculo várias variáveis

```python
x, y, z = sym.symbols('x y z')
f = x**2 + y**2 + z**2
f
```

```python
sym.diff(f, x), sym.diff(f, y), sym.diff(f, z)
```

```python
x, y = sym.symbols('x y')
f = Function('f')
u = f(x, y)
u
```

```python
ux = u.diff(x)
ux
```

```python
uy = u.diff(y)
uy
```

```python
eq = Eq(1 + (2*(ux/u)) + (3*(uy/u)), 0)
eq
```

```python
pdsolve(eq)
```

```python
x,y,z,c = sym.var('x , y , z, c')
u = Function('u')
w = Function('w')
# Definindo a função
f = u(x, y, z)
# Definindo as derivadas parciais
ux = f.diff(x)
uy = f.diff(y)
uz = f.diff(z)
# Definindo a PDE
eq = Eq(y * ux + z * uy + (c**2 - y - x**2/2) * uz, 0)
eq
```

```python
u = sym.exp(z*(c**2 - y - x**2/2))
u
```

```python
ux = u.diff(x)
ux
```

```python
uy = u.diff(y)
uy
```

```python
uz = u.diff(z)
uz
```

```python
expr = y * ux + z * uy + (c**2 - y - x**2/2) * uz
expr
```

```python
sym.simplify(expr)
```

# Lagrangeana

```python
t = sym.symbols('t')
x = sym.Function('x')(t)
L = sym.Rational(1,2)*sym.diff(x, t)**2 - x**2
L
```

```python
sym.diff(sym.diff(L, sym.diff(x, t)), t) - sym.diff(L, x)
```
