## Cardioid and parametric curves

```python
cardioid = (x**2 + y**2 - 2 * x)**2 - 4 * (x**2 + y**2)
cardioid
```

```python
t = np.linspace(0,2*np.pi,100)

x = 2 * np.cos(t) - np.cos(2*t)
y = 2 * np.sin(t) - np.sin(2*t)

# Plot line with RGB tuple (red=1, green=0.2, blue=0.5)
# and 20pt line width
plt.plot(x, y, label='cardioid curve')

# Add features to our figure
plt.axis()
plt.grid(True)
plt.show()
```

```python
t = np.linspace(0,2*np.pi,100)

x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

# Plot line with RGB tuple (red=1, green=0.2, blue=0.5)
# and 20pt line width
plt.plot(x,y,c=(1,0.2,0.5),lw=20)

# Add features to our figure
plt.title('Heart!')
plt.axis('equal')
plt.axis('off')

plt.show()
```

Check this [reference](https://elepa.files.wordpress.com/2013/11/fifty-famous-curves.pdf) to see fifty famous curves in cartesian plane and try to plot some of them.

```python
plt.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```
