```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import sympy
```

```python
import pylab as pylab

theta = pylab.linspace(0, 2.*pylab.pi, 1000)
a = 1.
r = 2 * a * (1. + pylab.cos(theta))
pylab.polar(theta, r)
pylab.show()
```

```python
T = np.linspace(0 , 2 * np.pi, 1024)
plt.axes(polar = True)
plt.plot(T, 1. + .25 * np.sin(16 * T), c= 'k')

plt.show()
```

```python
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(-np.pi,np.pi,1000)

# the function, which is y = sin(x) here
y = np.sin(x)**2 * np.sin(np.cos(x))**2 / np.cos(x)**2

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'b')

# show the plot
plt.show()
```

```python
def gain_dip(theta, phi):
    return 1.641*(np.cos(np.pi/2*np.cos(theta))/np.sin(theta))**2

theta = np.arange(-np.pi, np.pi,0.01)
# plot
ax = plt.subplot(111, polar=True)
# set zero west
ax.set_theta_zero_location('W')
ax.set_theta_direction('clockwise')
# let set an azimuth for example, pi
plt.plot(theta, gain_dip(theta, np.pi))
```

```python
phi = np.arange(-np.pi, np.pi,0.01)
ax = plt.subplot(111, polar=True)
# set zero west
ax.set_theta_zero_location('W')
ax.set_theta_direction('clockwise')
plt.plot(phi, gain_dip(np.pi/2, "not used")*np.ones_like(phi))
```

```python
theta = pylab.linspace(0, pylab.pi, 10000)
a = 1.5
r = a**2 * pylab.tan(theta)**2 * pylab.sin(pylab.pi * a * pylab.cos(theta))**2
pylab.polar(theta, r)
pylab.show()
```

```python
from matplotlib import pyplot as plt

I = 1
l = 1.5
l_lambda = 1
c = 1

text_legend = 'I =' + str(I) + ' l =' + str(l) + ', lambda =' + str(l_lambda) + ', c =' + str(c)

theta = np.linspace(0, np.pi,200)

r = I**2 * (l * np.pi/(c*l_lambda)) * np.tan(theta)**2 * np.sin(np.pi * l**2 * np.cos(theta)/l_lambda)**2

z = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize = (10,8))
plt.plot(y,z,'r', label = text_legend)
plt.xlabel('plano x-y', weight = 'bold', fontsize = 14)
plt.ylabel('z', weight = 'bold', fontsize = 14)
plt.xticks(weight = 'bold', fontsize = 14)
plt.yticks(weight = 'bold', fontsize = 14)
plt.legend()
plt.ylim([-0.1,0.1])
plt.xlim([-0.05,0.25])
plt.grid()
plt.show()
```

```python
I = 1
L = 1.5
lamb = 2
c = 1
const = I**2/2*np.pi*c
b = np.pi * L**2/lamb

theta = np.linspace(0, np.pi,200)
r = const * np.tan(theta)**2 * np.sin(b * np.cos(theta))**2
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.axhline(y=0, linestyle='-', color = 'black')
#plt.axis([0,4,-2,2])
plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```
